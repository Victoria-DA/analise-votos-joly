```python
import plotly.express as px

# Carrega os dados das eleições
df = px.data.election()

# Mostra os dados para conhecer as colunas
df.head()
```

```python
fig = px.bar(df, x='district', y='Joly', title='Votos do Joly por Distrito')
fig.show()
```

```python
fig2 = px.histogram(df, x='district', y='Joly', color='result', 
                   title='Votos do Joly por Distrito e Tipo de Vitória')
fig2.show()
```

### Análise das Eleições - Candidato Joly

**1. Maior quantidade de votos por região:**
De acordo com o gráfico de barras, a região que registrou a maior quantidade de votos para o candidato Joly foi **Côte-de-Liesse**, com um total de aproximadamente **3959** votos.

**2. Análise por tipo de vitória (Histograma):**
Ao analisar o histograma agrupado pelo tipo de resultado (`result`):
* No sistema de **Plurality**, a região com mais votos foi **Côte-de-Liesse**.
* No sistema de **Majority**, a região com mais votos foi **La Pointe-aux-Prairies**.
